# Fruits and Vegetables Prices Dataset

Product Name: Name of the produce item.
Date: The date of the pricing information.
Farm Price: The price at which the produce is sold at the farm.
Retail Prices: Retail prices in major cities (Atlanta, Chicago, Los Angeles, New York).
Average Spread: Percentage indicating the average markup between farm and retail prices.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
df=pd.read_csv('ProductPriceIndex.csv')
df.head()
```

```python
df.shape
```

```python
df.info() 
```

In this dataset we can observe that we dnt have any null values 

```python
df.describe(include = 'all')
```

```python
# Here we are changing date into the sequential format
df['date'] = pd.to_datetime(df['date'])
df.head() 
```

Now we will convert all the prices into numerical by removing the special character such as $, %.

```python
df['farmprice'] = pd.to_numeric(df['farmprice'].str.replace('[/$]','', regex=True))
df.head()
```

```python
df['atlantaretail'] = pd.to_numeric(df['atlantaretail'].replace('[/$]','', regex=True))
df.head()
```

```python
df['chicagoretail'] = pd.to_numeric(df['chicagoretail'].replace('[/$]','', regex=True))
df.head()
```

```python
df['losangelesretail'] = pd.to_numeric(df['losangelesretail'].replace('[/$]','', regex=True))
df.head()
```

```python
df['newyorkretail'] = pd.to_numeric(df['newyorkretail'].replace('[/$]','', regex=True))
df.head()
```

```python
df['averagespread'] = pd.to_numeric(df['averagespread'].replace('[/%,]','', regex=True))
df.head()
```

```python
df.dtypes
```

Now we are checking for the duplicate values and we are dropping them.

```python
df[df.duplicated(keep = False)].sort_values(['productname', 'date'])
```

```python
df = df.drop_duplicates(keep = 'first')
```

```python
df.shape
```

```python
sns.displot(df['averagespread'])
plt.show()
```

As we can observe from above chart the data we have is right skewed.

```python
Q1 = df['averagespread'].quantile(0.25)
Q3 = df['averagespread'].quantile(0.75)
IQR = Q3-Q1
ub = Q3+1.5*IQR
lb = Q1-1.5*IQR
df = df[(df['averagespread']>= lb) & (df['averagespread']<=ub) ]
df.shape
```

```python
sns.boxplot(df.averagespread)
plt.show()
```

```python
df1 = df.drop('productname', axis =1)
```

```python
sns.heatmap(df1.corr(), annot = True)
plt.show()
```

The heatmap of the correlation matrix reveals relationships between different variables. Notably, there is a correlation between farm price and average spread, suggesting that as farm prices increase, the spread tends to increase as well.

```python
farmprice_product_name = df.groupby('productname')['farmprice'].mean().reset_index().sort_values('farmprice', ascending = True)
farmprice_product_name
```

```python
sns.barplot(data = farmprice_product_name.head(5), x='farmprice', y='productname' )
plt.show()
```

```python
sns.barplot(data = farmprice_product_name.tail(5), x='farmprice', y='productname' )
plt.show()
```

Here in the above chart we are comparing the farmprice in X axis and product name in Y axis. From the above chart we can observe that oranges have low farmprice, and strawberies have high farmprice.
